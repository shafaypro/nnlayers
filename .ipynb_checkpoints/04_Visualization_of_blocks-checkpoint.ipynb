```python
import mxnet as mx
from mxnet import nd,viz
from mxnet.gluon import nn, model_zoo
```

### A Visual Geometry Group Metrics Import

```python
vgg11 = model_zoo.vision.vgg11(pretrained = True)  # Visual Geometry group pre build model
```

```python
vgg11 # Check out the shades babe
```

```python
# Getting the summary for the vgg
vgg11.summary(nd.ones((1,3,224,224))) # Gets the summary for the network based on the intermediate shapes 
```

#### Plotting your neural network of VGG11

```python
viz.plot_network(vgg11(mx.sym.var(name ='data')), shape={'data':(1,3,224,224)})
```

### Checking out the mobile net

```python
mobile_net = model_zoo.vision.mobilenet_v2_1_0(pretrained=True)
```

```python
mobile_net
```

```python
mobile_net.summary(nd.ones((1,3,64,64)))
```
